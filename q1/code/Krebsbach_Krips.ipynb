## SELF DISCLOSURE

```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
import itertools
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/pearlvishen/Downloads/Depth-of-Self-Disclosure.csv")

rater_columns = [
    'Gemini 2.5', 
    'Claude 3.7', 
    'GPT 4o'
]

kappa_scores = {}
for r1, r2 in itertools.combinations(rater_columns, 2):
    kappa = cohen_kappa_score(df[r1], df[r2], weights='quadratic')
    kappa_scores[(r1, r2)] = kappa

kappa_matrix = pd.DataFrame(index=rater_columns, columns=rater_columns, dtype=float)

for (r1, r2), score in kappa_scores.items():
    kappa_matrix.loc[r1, r2] = score
    kappa_matrix.loc[r2, r1] = score
    kappa_matrix.loc[r1, r1] = 1.0
    kappa_matrix.loc[r2, r2] = 1.0

print("Quadratically Weighted Cohen's Kappa Scores:")
print(kappa_matrix.round(3))

plt.figure(figsize=(10, 8))
sns.heatmap(kappa_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title("Pairwise Quadratic Weighted Kappa Between Raters")
plt.tight_layout()
plt.show()

#Kripendorff's:
import pandas as pd
import krippendorff


ratings = df[rater_columns].transpose()

alpha = krippendorff.alpha(reliability_data=ratings.values, level_of_measurement='ordinal')

print(f"Krippendorff’s alpha (ordinal) for the group of raters: {alpha:.3f}")
```
